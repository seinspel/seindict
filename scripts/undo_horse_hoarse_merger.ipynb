```python
from utils import apply, FindAndReplace, Replacement
```

```python
spelling_patterns = [
    # "(?<=[TDN])U(?![IE])",
    "OAR",
    "uR",
    "ORE(?![AEIOUauie])",
    "(?<=[^AEIOUauie][BFV])OR(?![M])",
    "(?<=^[BFV])OR(?![M])",
    "(?<=[^SAEIOUauie]P)OR",
#     "OR(?=[^AEIOUauie]E$)", <- this one is useless and only causes problems
#     "OA(?=R[AEIOUaY])",  # two R followed by a vowel
    # "(?<=[TDN])EU",
]
phoneme_patterns = [
    Replacement(
        # guard=f"OR{stress}(?= R [AEIOU])",
        guard=f"OR{stress}(?! [AEIOUə])",
        source=f"OR{stress}",
        target=f"OHR{stress}"
    )
    for stress in (0, 1, 2)
]
```

```python
overrides = {
}

# these are with NORTH vowel
exclude = """
border
born
California
corpse
cyborg
dearborn
endorse
firstborn
for
FORBEARANCE
fork
forklift
form
fortress
forty
freeborn
george
HAWTHORNE
HORSE
HORSEBACK
HORSED
HORSEMAN
HORSEMANSHIP
HORSEMEN
HORSEPLAY
HORSEPOWER
HORSERADISH
HORSES
HORSESHIT
HORSESHOE
HORSETAIL
important
importance
inborn
morgue
morning
morse
morsel
newborn
newborns
norse
pitchfork
porn
reborn
racehorse
racehorses
spork
stillborn
unborn
warn
wellborn
WHITEHORSE
WORKHORSE
WORKHORSES
""".upper()
```

```python
find_long_or = FindAndReplace(
    spelling_patterns=spelling_patterns,
    phoneme_patterns=phoneme_patterns,
    overrides=overrides,
    exclude=exclude,
)

for_real = False
find_long_or.is_active = for_real
apply(
    find_long_or,
    save_result=for_real,
    only_first_file=False,
    mode="transform",
)
```
