```python
spelling_patterns = [
#     "(?<=[^AEIOUauie])LE$",
#     "(?<=[^AEIOUauie])L(?=ING$)",
#     "(?<=[^AEIOUauie])LE(?![AEIOUauie])",
    "[AE]L"
]
phoneme_patterns = [
    Replacement(
        # guard=f"OR{stress}(?= R [AEIOU])",
        guard=f"(?<![012]) ə L(?! [AEIOUə])",
        source=f"ə L",
        target=f"EL"
    )
#     for stress in (0, 1, 2)
]
```

```python
overrides = {
}

# these are with NORTH vowel
exclude = """

""".upper()
```

```python
find_long_or = FindAndReplace(
    spelling_patterns=spelling_patterns,
    phoneme_patterns=phoneme_patterns,
    overrides=overrides,
    exclude=exclude,
    leeway=2,
)

for_real = True
find_long_or.is_active = for_real
apply(
    find_long_or,
    save_result=for_real,
    only_first_file=False,
    mode="transform",
)
```
