```python
from dataclasses import dataclass
from utils import apply, FindAndReplace, Replacement, is_vowel
```

```python
def is_open_syllable(pronun, index):
    if len(pronun) == index + 1:
        return False
    return is_vowel(pronun[index + 1])
```

```python
@dataclass
class Occurrances:
    phonemes: list
    in_open_syllable: bool = False
    
    def __call__(self, word, pronun, dictionary):
        pronun = pronun.split(" ")
        for i, phoneme in enumerate(pronun):
            if phoneme in self.phonemes:
                is_occurrance = (not self.in_open_syllable) or is_open_syllable(pronun, i + 1)
                # confusingly, we have to return `False` when we found something
                return not is_occurrance
        return True
```

```python
apply(
    Occurrances(["A0", "A2"], in_open_syllable=True),
    save_result=False,
    only_first_file=False,
    mode="filter",
)
```
