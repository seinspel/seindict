```python
from utils import apply, FindAndReplace, Replacement
```

```python
spelling_patterns = [
    # "(?<=[TDN])U(?![IE])",
    # "IR",
    "ER",
    "ɛR",
]
phoneme_patterns = [
    Replacement(
        # guard=f"OR{stress}(?= R [AEIOU])",
        guard=f"UR{stress}",
        source=f"UR{stress}",
        target=f"ER{stress}"
    )
    for stress in (0, 1, 2)
]
```

```python
overrides = {
}

exclude = """
""".upper()
```

```python
find_long_or = FindAndReplace(
    spelling_patterns=spelling_patterns,
    phoneme_patterns=phoneme_patterns,
    overrides=overrides,
    exclude=exclude,
)

for_real = True
find_long_or.is_active = for_real
apply(
    find_long_or,
    save_result=for_real,
    only_first_file=False,
    mode="transform",
)
```
