```python
import re
from utils import apply
```

```python
exclude = """
COWHER
DEWHIRST
DEWHURST
LAWHORN
""".split()


def find_wh_words(word, pronun, dictionary):
    if word in exclude:
        return True
    
    # replace digraphs with single letters so that the index calculation is more reliable
    word_with_better_letters = word.replace("SH", "ʃ").replace("TH", "θ").replace("WH", "ʍ").replace("CH", "ʒ")
    
    wh_indices = [m.start() for m in re.finditer("ʍ", word_with_better_letters)]        
    if not wh_indices:  # no suitable letters found
        return True

    pronun_l = pronun.split()
    w_indices = [i for i, pronun_el in enumerate(pronun_l) if pronun_el == "W"]
    if not w_indices:
        return True

    leeway = 2 if len(w_indices) == 1 else 1  # if there is more than one W, don't get them mixed up

    changed_anything = False
    for w_index in w_indices:
        if min(abs(wh_index - w_index) for wh_index in wh_indices) > leeway:
            print(f"probably no candiate '{word}', letter indices: {wh_indices}, sound index: {w_index}")
            continue
        pronun_l[w_index] = "WH"
        changed_anything = True
    if not changed_anything:
        return True
    print(f"change '{word.lower().replace('wh', 'WH')}', new pronun: '{' '.join(pronun_l).lower().replace('wh', 'WH')}'")
    return " ".join(pronun_l)
```

```python
apply(
    find_wh_words,
    save_result=True,
    only_first_file=False,
    mode="transform",
)
```
